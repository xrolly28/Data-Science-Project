```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import seaborn as sns
df=pd.read_csv('/content/top100companies.csv')
```

```python
import nsepy
from datetime import date , timedelta
data_s=nsepy.get_history('BSE',start=date.today()-timedelta(days=100),end=date.today()-timedelta(days=0))
data_s=data_s.drop(data_s.columns.difference(['VWAP']),axis=1)
data_s=data_s.rename(columns={"VWAP": "Sensex"})
print(data_s)
```

```python
df1=pd.DataFrame()
for i in range(100):
  symbol=df['Symbol'][i]
  data_n=nsepy.get_history(symbol,start=date.today()-timedelta(days=100),end=date.today()-timedelta(days=0))
  data_n=data_n.drop(data_n.columns.difference(['VWAP']),axis=1)
  data_n=data_n.rename(columns={"VWAP": symbol})
  df1=pd.concat([df1,data_n],axis=1)
```

```python
data=pd.concat([df1,data_s],axis=1)
print(data)
```

```python
data.to_csv('companies_VWAP_data.csv', index = False)
from google.colab import files
files.download("companies_VWAP_data.csv")
```

```python
data=data.dropna(axis=1)
# data.isna
data.head()
```

```python
from sklearn.model_selection import train_test_split
features=data.columns.difference(['Sensex'])
train_x, test_x, train_y, test_y=train_test_split(data[features],data['Sensex'],test_size=0.3,random_state=123)

import statsmodels.api as sm
train_x= sm.add_constant(train_x)
lm=sm.OLS(train_y,train_x).fit()
print(lm.summary())
```

those companies which is having higher magnitude coefficient are contributing in sensex index.

```python
print('Parameters: ', lm.params)
print('R2: ', lm.rsquared)
```

```python
test_x = sm.add_constant(test_x)
y_pred = lm.predict(test_x)
 
from sklearn import metrics
import numpy as np

print ('MAE:', metrics.mean_absolute_error(test_y, y_pred))
print ('MSE:', metrics.mean_squared_error(test_y, y_pred))
print ('RMSE:', np.sqrt(metrics.mean_squared_error(test_y, y_pred)))
```
